```python
import selenium
from selenium import webdriver as wb
webD = wb.Chrome("chromedriver.exe")
webD.get('https://webscraper.io/test-sites/e-commerce/allinone')
```

```python
webD.find_element_by_xpath('//*[@id="side-menu"]/li[2]/a').click()
```

```python
webD.find_element_by_xpath('//*[@id="side-menu"]/li[2]/ul/li[1]/a').click()
```

```python
productInfoList = webD.find_elements_by_class_name('thumbnail')

listOfLinks = []

for el in productInfoList:
    ppp = el.find_elements_by_tag_name('h4')[-1].find_element_by_tag_name('a').get_property('href')
    listOfLinks.append(ppp)
```

```python
from tqdm import tqdm

allDetails = []

for i in tqdm(listOfLinks):
    webD.get(i)
    name = webD.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[1]/h4[2]').text
    description = webD.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[1]/p').text
    price = webD.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[1]/h4[1]').text
    review = webD.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[3]/p').text
    tempJ = {
        'name': name,
        'price': price,
        'review': review,
        'description': description,
        'link': i
    }
    allDetails.append(tempJ)
```

```python
import pandas as pd
pd.DataFrame(allDetails)
```
